# Ripple Demo

This notebook demonstrates **jupyterlab-ripple**, which adds reactive execution to JupyterLab.

**To get started:** click the **Ripple** toggle button in the toolbar above to enable reactive mode.

Once enabled, executing a cell will automatically re-execute all downstream cells that depend on its variables.

---
## 1. Basic Reactivity

Run the cells below in order. Then change `x` in the first cell and re-run it — watch `y` and `result` update automatically.

```python
x = 100000
```

```python
y = x * 2
print(f"y = x * 2 = {y}")
```

```python
result = y + 10
print(f"result = y + 10 = {result}")
```

**Try it:** Change `x = 5` to `x = 100` above and re-run that cell. Both `y` and `result` update in one shot.

---
## 2. Branching Dependencies

One cell can feed into multiple independent downstream branches. Changing `base` triggers both branches.

```python
base = 3
```

```python
squared = base ** 2
print(f"squared = {base}² = {squared}")
```

```python
cubed = base ** 3
print(f"cubed = {base}³ = {cubed}")
```

```python
total = squared + cubed
print(f"total = {squared} + {cubed} = {total}")
```

**Try it:** Change `base = 3` to `base = 7`. All three downstream cells re-execute in the correct order.

---
## 3. Functions and Imports

Ripple tracks function definitions and import statements. Re-defining a function re-runs cells that call it.

```python
import math
```

```python
def circle_area(radius):
    return math.pi * radius ** 2
```

```python
r = 10
```

```python
area = circle_area(r)
print(f"Area of circle with radius {r}: {area:.2f}")
```

**Try it:** Change `r = 5` to `r = 10` and re-run — the area updates. Or change the function body (e.g. return the circumference instead) and re-run — the output cell also re-executes.

## 4. Data Pipeline

A more realistic scenario: a small data processing pipeline where changing the raw data or a parameter propagates through all downstream stages.

```python
# Raw data
temperatures_f = [32, 68, 72, 85, 90, 55, 60, 78]
```

```python
# Convert to Celsius
temperatures_c = [(f - 32) * 5 / 9 for f in temperatures_f]
print("Celsius:", [f"{t:.1f}" for t in temperatures_c])
```

```python
# Summary statistics
avg_temp = sum(temperatures_c) / len(temperatures_c)
min_temp = min(temperatures_c)
max_temp = max(temperatures_c)
print(f"Average: {avg_temp:.1f}°C")
print(f"Range: {min_temp:.1f}°C to {max_temp:.1f}°C")
```

```python
# Classification
cold = [t for t in temperatures_c if t < 15]
warm = [t for t in temperatures_c if 15 <= t < 25]
hot = [t for t in temperatures_c if t >= 25]
print(f"Cold (<15°C): {len(cold)} readings")
print(f"Warm (15-25°C): {len(warm)} readings")
print(f"Hot (>=25°C): {len(hot)} readings")
```

**Try it:** Add or remove values from `temperatures_f` and re-run that cell. The conversion, statistics, and classification all update automatically.

## 5. Visual Indicators

With reactive mode enabled, notice the colored left borders on cells:

| Color | Meaning |
|-------|-------|
| **Blue** | Upstream — this cell defines variables used by others |
| **Green** | Downstream — this cell depends on variables from others |
| **Amber** | Stale — an upstream cell changed but this hasn't re-run yet |
| **Red** | Conflict or cycle detected |

Hover over any cell to see its tooltip showing which variables it defines, references, and how many cells depend on it.

## 6. Independent Cells Are Unaffected

Cells that share no variables with the rest of the notebook are never auto-triggered.

```python
# This cell is independent — it won't re-run when you change cells above
import random
dice = random.randint(1, 6)
print(f"Dice roll: {dice}")
```

Re-run any cell in sections 1-4 above. This dice roll stays the same because it shares no variables with them.

## 7. Toggling Reactive Mode

Click the **Ripple** toolbar button to toggle reactive mode on and off at any time.

- **On**: colored borders appear; executing a cell triggers downstream re-execution
- **Off**: standard JupyterLab behavior; cells only run when you explicitly execute them

The toggle is per-notebook, so you can have reactive mode on in this notebook while another notebook behaves normally.
